```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Task 1, 2 Load the CSV file into a DataFrame named df and describe it
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
df = pd.read_csv('./files_for_lab/csv_files/marketing_customer_analysis.csv')
df.head()
```

```python

# Task 3: Show a plot of the total number of responses
# plt.figure(figsize=(8, 5))
df['Response'].value_counts().plot(kind='bar', title='Total Number of Responses')
plt.ylabel('Count')
plt.xlabel('Response Type')
plt.show()

# Task 4: Show a plot of the response rate by the sales channel
response_rate_by_channel = df.groupby('Sales Channel')['Response'].apply(lambda x: (x == 'Yes').mean())
response_rate_by_channel.plot(kind='bar', title='Response Rate by Sales Channel')
plt.ylabel('Response Rate')
plt.xlabel('Sales Channel')
plt.show()

# Task 5: Show a plot of the response rate by the total claim amount
# Binning the total claim amount
df['Total Claim Amount Bins'] = pd.qcut(df['Total Claim Amount'], q=10)
df.groupby('Total Claim Amount Bins')['Response'].apply(lambda x : (x == 'Yes').mean()).plot(kind = 'bar', title='Response Rate by Total Claim Amount')
plt.ylabel('Response Rate')
plt.xlabel('Total Claim Amount Bins')
plt.xticks(rotation=45)
plt.show()
'''
# Task 6: Show a plot of the response rate by income
# Binning the income
df['Income Bins'] = pd.qcut(df['Income'], q=10)
response_rate_by_income = df.groupby('Income Bins')['Response'].apply(lambda x: (x == 'Yes').mean())
response_rate_by_income.plot(kind='bar', title='Response Rate by Income')
plt.ylabel('Response Rate')
plt.xlabel('Income Bins')
plt.xticks(rotation=45)
plt.show()
'''
# Task 6: Show a plot of the response rate by income (corrected binning)
# Binning the income
df['Income Bins'] = pd.qcut(df['Income'], q=10, duplicates='drop')
response_rate_by_income = df.groupby('Income Bins')['Response'].apply(lambda x: (x == 'Yes').mean())
response_rate_by_income.plot(kind='bar', title='Response Rate by Income')
plt.ylabel('Response Rate')
plt.xlabel('Income Bins')
plt.xticks(rotation=45)
plt.show()
```
